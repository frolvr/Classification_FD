```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
from sklearn.linear_model import LogisticRegression

# предобработка
REPLACER = {' г.': ' г', 
            ' р.': ' р',
            '?': '? ',
            '!': '! '}
DELIMIT = '. ', '? ', '! '

# порог
THRESHOLD = 0.3

# данные
TRAIN_FILE = 'БК База знаний.xlsx'
TRAIN_X = 'Вопрос'
TRAIN_Y = 'Тип-подкатегория-подподкатегория'

TEST_FILE = '01.04-31.08-18636-no-dubl.csv'
TEST_X = 'Описание'
TEST_Y = 'Теги'
TEST_SCORE = 'Score'

def splitter(delimiters, string, maxsplit=0):
    string = string.lower()
    string = string.lstrip()
    for key, value in REPLACER.items():
        string = string.replace(key, value)
    
    regexPattern = '|'.join(map(re.escape, delimiters))
    
    # защита от бессмысленных строк
    if any(x in string for x in 'абвгдеёжзийклмнопрстуфхцчшщъьэюя'):
        results = re.split(regexPattern, string, maxsplit)
        return list(filter(None, results))
    else:
        return np.NaN


def prediction(dframe):
    result = ''
    scores = []
    # print(dframe['text'])
    response = vectorizer.transform(dframe['text'])
    
    result = lr.predict(response)
    scores = np.max(lr.predict_proba(response), axis=1)
    result = list(set(result))

    dframe[TEST_Y] = "|".join(result)
    dframe[TEST_SCORE] = min(scores)
    return dframe


# загрузка
print('Загрузка данных ... ', end='')
train_data = pd.read_excel(TRAIN_FILE, header=0)
test_data = pd.read_csv(TEST_FILE, header=0)
print('выполнена')


# 0. Провести предобработку данных
print('Предобработка ... ', end='')
train_data = train_data.dropna(subset=[TRAIN_Y])
test_data = test_data.dropna(subset=[TEST_X])

# 1. Разбить описание на предложения
test_data['text'] = test_data[TEST_X].apply(lambda x:splitter(DELIMIT, x))
test_data = test_data.dropna(subset=['text'])
print('закончена')
#print(test_data['text'][1])

# заполнение векторизатора
print('Обучение модели ... ', end='')
vectorizer = TfidfVectorizer(min_df=2, max_df=0.75, max_features=20000, smooth_idf=True, use_idf=True, sublinear_tf=False, norm='l2')
word_count_vector=vectorizer.fit_transform(train_data[TRAIN_X], train_data[TRAIN_Y])

lr = LogisticRegression(solver='lbfgs', multi_class='multinomial')
lr.fit(word_count_vector, train_data[TRAIN_Y])
print('выполнено')

# магия
print('Классификация ... ', end='')
test_data = test_data.apply(prediction, axis=1)
print('закончена')

# удаление вспомогательных колонок
test_data = test_data.drop(['text'], axis=1)
# сохранение результатов
print('Сохранение данных ... ', end='')
clean = test_data[test_data[TEST_SCORE] >= THRESHOLD]
clean.to_csv('clean.csv')
basket = test_data[test_data[TEST_SCORE] < THRESHOLD]
basket.to_csv('basket.csv')
print('закончено')


```
